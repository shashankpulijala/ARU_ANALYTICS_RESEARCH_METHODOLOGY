```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

titanic = pd.read_csv('./titanic/train.csv')
```

```python
titanic.head()
```

```python
titanic.describe()
```

# AGE Distribution

```python
plt.hist(titanic['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
#Normal distribution
```

# Gender 

```python
sns.barplot(x='Sex', y='Survived', data=titanic)
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Correlation Matrix of different numerical variables
```

```python
corr = titanic.corr()
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare vs. Age')
plt.show()
#higher the fare higher the chances of survival
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Sex', data=titanic)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare vs. Age')
plt.show()
```

```python
import seaborn as sns
# Create a barplot
sns.barplot(x="Pclass", y="Survived", data=titanic, estimator=sum)

# Add labels and title
plt.xlabel("Passenger Class")
plt.ylabel("Number of Survivors")
plt.title("Number of Survivors by Passenger Class")
```

```python
pclass_survival = titanic.pivot_table(values='PassengerId', index='Survived', columns='Pclass', aggfunc='count')

# Create a stacked bar chart
pclass_survival.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.title('Number of Survivors and Non-Survivors by Passenger Class')
```

```python
#clearly pclass 1 has better survival rate than the pclass 3
```
